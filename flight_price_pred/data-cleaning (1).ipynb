## 5.Cleaning operation

```python
def convert_to_minutes(ser):
    return(
        ser
            .str.split(" ",expand=True) ## expand=True will give dataframe
            .set_axis(["hour","minute"],axis=1) # change the name of columns
            .assign(
             hour=lambda df_:(
                df_
                .hour
                .str.replace("h","")
                .astype(int)
                .mul(60)
               )
              ,
              minute=lambda df_:(
                df_.minute
                .str.replace("m","")
                .fillna("0")
                .astype(int)
               )
              )
              .sum(axis=1) ##This will sum the values for each row  as axis =1
              )
```

```python
def clean_data(df):
    return (
        df
        .drop(index=[6474])
        .drop_duplicates()
        .assign(
            #airline=lambda df_:df_.airline.str.lower()
             **{
                 col: df[col].str.strip()
             for col in df.select_dtypes(include="O").columns
             }
           
        )
        .rename(columns=str.lower)
        .assign(
            airline=lambda df_:(
                df_
                .airline
                .str.replace(" Premium economy", "")
                .str.replace(" Business", "")
                .str.title()
            ),
            date_of_journey=lambda df_: pd.to_datetime(df_.date_of_journey,format='%d/%m/%Y')
            ,
            destination= lambda df_:(
                df_
                .destination
                .str.replace("Banglore","Bangalore")
                .str.replace("Cochin", "Kochi")
            ),
            dep_time=lambda df_:pd.to_datetime(df_.dep_time).dt.time,
            #arrival_time = lambda df_: df_.arrival_time.loc[lambda ser: ser.str.contains("[^0-9:]")]
            #.str.split(" ").str[0]
            arrival_time=lambda df_:pd.to_datetime(df_.arrival_time).dt.time,
            #method-1---> duration=lambda df_: convert_to_minutes(df_.duration)
            #method-2:
            duration=lambda df_:df_.duration.pipe(convert_to_minutes) # here we simply pass the df_.duration columns to "convert_to_minutes" function
            ,
            total_stops=lambda df_:(
                df_
                .total_stops
                .replace("non-stop","0")
                .str.replace(" stops?", "", regex=True) # " stop" " stops"
                #.astype(int) is not work here as there is mising value in total_stops columns ans nan is float bydefaultt
                .pipe(lambda ser: pd.to_numeric(ser)) # it compatible with float beacuse nan is also float so, it give output in float format
        
            )
            ,
            additional_info=lambda df_: df_.additional_info.replace("No info","No Info")
        
            
        )
        .drop(columns="route")
        
        
       
    )
```

```python
clean_data(flight).additional_info.unique()
```

```python
flight_cleaned = clean_data(flight)
flight_cleaned
```
